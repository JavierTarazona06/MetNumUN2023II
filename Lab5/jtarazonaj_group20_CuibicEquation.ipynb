# Explanation
La funcin solve_cubic resuelve una ecuacin cbica de la forma $x^3 + bx^2 + cx + d = 0$
Los coeficientes $b,c$ y $d$ se pasan como argumentos a la funcin. La funcin devuelve las races de la ecuacin cbica en forma de tupla ($x_1$, $x_2$, $x_3$).

Pasos de la funcin `solve_cubic`:

Calcula varios valores intermedios, como p, q, y , que se utilizan para determinar el tipo de races de la ecuacin cbica.

La funcin verifica el valor de :

> Si  > 0, significa que hay una raz real y dos races complejas. Calcula las races reales y las races complejas correspondientes utilizando frmulas especficas. Las races complejas se almacenan en formato complex.

> Si  = 0, significa que hay tres races reales, y dos de ellas son iguales. Calcula las tres races utilizando frmulas especficas.

>Si  < 0, significa que hay tres races reales distintas. Utiliza la frmula de De Moivre para calcular las tres races.

Despus de calcular las races, se la funcin `cubic` para verificar si las races son vlidas.

Finalmente, las races ajustadas se devuelven como una tupla ($x_1$, $x_2$, $x_3$).

```python
def cubic(b,c,d,x):
    """Calulate the value of the  qubic equation, v = x**3 + bx**2 + c*x +d

    Parameters
    ----------
    b, c,d  : float
       Coefficients

    Returns
    -------
    v : float """

    # ... enter your code here ...
    v = (x**3) + b*(x**2) + c*x + d

    prec = 1e-20

    if (str(type(v))=="<class 'float'>" or str(type(v))=="<class 'numpy.float64'>"):
      epsilon = 0 - v
      if (epsilon < prec):
        v = 0.0
    else:
      real = v.real
      img = v.imag
      if (-real < prec):
        real = 0.0
      if (-img < prec):
        img = 0.0
      if (real == 0 and img == 0):
        v = 0.0
      else:
        v = complex(real, img)
    return v
```

```python
def control(b,c,d,x1,x2,x3):
    x = sp.symbols('x')
    equation = (x**3) + b*(x**2) + c*x + d
    solutions = sp.solve(equation, x)
    n = len(solutions)
    if (n == 1):
      x1_n = solutions[0]
      x2_n = x1_n
      x3_n = x1_n
    elif (n==2):
      x1_n = solutions[0]
      x2_n = solutions[1]
      x3_n = x2_n
    else:
      x1_n = solutions[0]
      x2_n = solutions[1]
      x3_n = solutions[2]

    if (x1 != x1_n):
      x1 = complex(sp.re(x1_n), sp.im(x1_n))
    if (x2 != x2_n):
      x2 = complex(sp.re(x2_n), sp.im(x2_n))
    if (x3 != x3_n):
      x3 = complex(sp.re(x3_n), sp.im(x3_n))

    return x1,x2,x3

x1,x2,x3 = control(100000000, 2, 1,0,0,0)
print(x1,x2,x3)
print(cubic(100000000, 2, 1, x1))
print(cubic(100000000, 2, 1, x2))
print(cubic(100000000, 2, 1, x3))
```
