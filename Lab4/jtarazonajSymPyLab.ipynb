My own example: x = I*x1 + 2*I*x2,y = 5*y1 - 7*I*y2, z = -4*z1 + 4*I*z2, x*y*z

image.png

```
from sympy import symbols, I, N, pi, expand
x1, x2, y1, y2, z1, z2 = symbols("x1 x2 y1 y2 z1 z2", real=True)

x = I*x1 + 2*I*x2
y = 5*y1 - 7*I*y2
z = -4*z1 + 4*I*z2

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
w=1/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
```
